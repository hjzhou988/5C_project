```python
# Fragment ('FRF' or 'RFR') length distrution
# open file called 'xxxx_loops_length_and_coordinates.txt'
# 3,4 and 7,8 column are the coordinates of the range of each fragment
import matplotlib.pyplot as plt
import numpy as np
def collectFragmentLength(File):  # a function that reads the file and return all fragment lengths in that file
    Fraglength=[]
    with open(File,'r') as f:
        for line in f:
            tmp=line.strip().split()
            Fraglength.append(int(tmp[3])-int(tmp[2]))
            Fraglength.append(int(tmp[7])-int(tmp[6]))
    return Fraglength
```

```python
Constitutive=collectFragmentLength('constitutive_loops_length_and_coordinates.txt')
plt.hist(Constitutive,50)
plt.title('constitutive_loops_fragment_length_distribution')
plt.show()
```

```python
All=collectFragmentLength('all_loops_length_and_coordinates.txt')
plt.hist(All,50)
plt.title('All_loops_fragment_length_distribution')
plt.show()
```

```python
Nonsig=collectFragmentLength('non-significant_loops_length_and_coordinates.txt')
plt.hist(Nonsig,50)
plt.title('non-significant_loops_fragment_length_distribution')
plt.show()
```

```python
ES=collectFragmentLength('mES_specific_loops_length_and_coordinates.txt')
plt.hist(ES,50)
plt.title('ES_specific_loops_fragment_length_distribution')
plt.show()
```
